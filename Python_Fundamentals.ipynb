Functions

```python
def happy_birthday(name, age):
  print(f"happy birthday {name}")
  print(f"Party chayo {age} ko budo")
  print("kaile dinchas party")
  print("Happy birthday to you")
happy_birthday("atik", 22)
happy_birthday("broman", 22)
```

```python
def display_invoice(username, amount, due_date):
  print(f"hello {username}")
  print(f"Your total amount is Rs.{amount}")
  print(f"Your dead line is on {due_date}")

display_invoice("aeek", 200000000, "12/12/2025")
```

```python
def add(x,y):
  z = x + y
  return z

def subtract(x,y):
  z = x - y
  return z

print(add(2, 3))
```

```python
def create_name(first, last):
  first = first.capitalize()
  last = last.capitalize()
  return first + " " + last

full_name = create_name("satik", "sahukhala")
print(full_name)
```
